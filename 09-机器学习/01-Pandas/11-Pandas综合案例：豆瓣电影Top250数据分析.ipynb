```python
import pandas as pd
```

```python
df = pd.read_excel('./movie_top_250.xlsx')
```

```python
df
```

```python
# 分析：把评价最多的电影使用WordCount词云显示
```

```python
# 数据变量定义
title = df.title.to_list()
# 对nums数据清洗
nums = df.nums.str.rstrip('人评价').to_list()
```

```python
# 把title和nums合并为[(), (), ()]，也可以是[[], [], []]
data = list(zip(title, nums))
```

```python
# 具体词云实现
import pyecharts.options as opts
from pyecharts.charts import WordCloud

c = (
    WordCloud()
    .add(series_name="豆瓣电影评价词云分析", data_pair=data, word_size_range=[6, 66])
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title="豆瓣电影评价词云分析", title_textstyle_opts=opts.TextStyleOpts(font_size=23)
        ),
        tooltip_opts=opts.TooltipOpts(is_show=True),
    )
)
c.render_notebook()
```

```python
# 豆瓣电影评分分布情况 => 9.7分电影有多少部 => count(电影) group by 评分
```

```python
# 准备数据
# 评分信息
df.groupby('grade')['title'].count()
# 每个评分电影数量
x_data = df.groupby('grade')['title'].count().index.to_list()
y_data = df.groupby('grade')['title'].count().to_list()
```

```python
from pyecharts import options as opts
from pyecharts.charts import Bar
from pyecharts.faker import Faker


c = (
    Bar()
    .add_xaxis(x_data)
    .add_yaxis("电影数量", y_data)
    .set_global_opts(title_opts=opts.TitleOpts(title="豆瓣电影评分分布情况"))
)
c.render_notebook()
```
