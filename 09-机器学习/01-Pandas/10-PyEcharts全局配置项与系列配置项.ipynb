```python
from pyecharts.charts import Map
from pyecharts.faker import Faker
# 配置项必须导入包
from pyecharts import options as opts
```

```python
c = (
    Map()
    .add("商家A", [list(z) for z in zip(Faker.provinces, Faker.values())], "china")
    .set_global_opts(title_opts=opts.TitleOpts(title="Map-基本示例"))  # 全局配置
)
c.render_notebook()
```

```python
# 标题配置
from pyecharts import options as opts
from pyecharts.charts import Gauge

c = (
    Gauge()
    .add("", [("完成率", 66.6)])
    .set_global_opts(
        # Notebook快捷键 => Ctrl + Q查帮助，注意：鼠标必须放置在圆括号的里面
        title_opts=opts.TitleOpts(title='仪表盘', title_link='https://pyecharts.org/', subtitle='传智教育提供技术支持！'),
        legend_opts=opts.LegendOpts(is_show=True),
        toolbox_opts=opts.ToolboxOpts(is_show=True),
        visualmap_opts=opts.VisualMapOpts(is_show=True)
    )
)
c.render_notebook()
```

```python
# 系列配置项 => .set_series_opts()，特别注意：系列配置项与图形有关，有些图形可以配置系列项，有些无法配置，具体参考example
```

```python
from pyecharts import options as opts
from pyecharts.charts import Bar
from pyecharts.faker import Faker

c = (
    Bar()
    .add_xaxis(Faker.choose())
    .add_yaxis("商家A", Faker.values())
    .add_yaxis("商家B", Faker.values())
    .reversal_axis()
    .set_series_opts(
        label_opts=opts.LabelOpts(position="right")
    )
    .set_global_opts(title_opts=opts.TitleOpts(title="Bar-翻转 XY 轴"))
)
c.render_notebook()
```

```python
from pyecharts import options as opts
from pyecharts.charts import Pie
from pyecharts.faker import Faker

c = (
    Pie()
    .add("", [list(z) for z in zip(Faker.choose(), Faker.values())])
    .set_global_opts(title_opts=opts.TitleOpts(title="Pie-基本示例"))
    .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {d}"))
)
c.render_notebook()

# 说明：
# 折线（区域）图、柱状（条形）图、K线图 : {a}（系列名称），{b}（类目值），{c}（数值）, {d}（无）
# 散点图（气泡）图 : {a}（系列名称），{b}（数据名称），{c}（数值数组）, {d}（无）
# 地图 : {a}（系列名称），{b}（区域名称），{c}（合并数值）, {d}（无）
# 饼图、仪表盘、漏斗图: {a}（系列名称），{b}（数据项名称），{c}（数值）, {d}（百分比）
```
