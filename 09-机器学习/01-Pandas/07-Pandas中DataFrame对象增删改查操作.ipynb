```python
# 在Pandas中不仅可以通过直接赋值添加列，还可以通过assign()方法实现
```

```python
df2
```

```python
df2.assign(new0=666)
```

```python
df2.assign(new1=[11, 22, 33, 44, 55])
```

```python
df2.assign(new2=pd.Series([111, 222, 333, 444, 555]))
```

```python
df2.assign(new3=df2.year + df2.GDP)
```

```python
def foo(df):
    # 当df2.assign()执行时，系统会自动将df2对象作为参数传给foo的第一个参数
    value = df.index.values
    return value
```

```python
df2.assign(new4=foo)
```

```python
df2.assign(
    new0=666,
    new1=[11,22,33,44,55],
    new2=pd.Series([111,222,333,444,555]),
    new3=df2.year * 2,
    new4=foo
)
```

```python
df7 = pd.DataFrame({
    '姓名':['小明', '小美', '小强', '小兰'],
    '成绩':[100, 90, 90, 80]
})
```

```python
df7
```

```python
df7.rank()
```

```python
df7.rank(method='average', ascending=True)
```

```python
df7['成绩排名'] = df7.成绩.rank(ascending=False)
df7
```

```python
df7['成绩排名'] = df7.成绩.rank(method='min', ascending=False)
df7
```

```python
df7['成绩排名'] = df7.成绩.rank(method='max', ascending=False)
df7
```

```python
df7['成绩排名'] = df7.成绩.rank(method='dense', ascending=False)
df7
```
